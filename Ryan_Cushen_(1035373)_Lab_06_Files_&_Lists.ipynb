## **Part A** - Creating a file
---

Here we will be retrieving data from the user input, adding it to a list, and then publishing it to a file so that it can persist after the program has been exited.<br/><br/>

### Your Task

Create a basic loop that reads in scores from the user until the user enters the sentinel value of -1. Each score should be added to a list. <br/><br/>

After the loop has been exited, the program should take the top 10 scores and print them in order (highest to lowest) to the screen, and also to a file called `highscores.txt` and then the program should exit.

<br/>


### Your Answer

```python
# Student Name: Ryan Cushen
# Student Number: 1035373
def main():
    # Initialize empty list to store scores
    scores = []

    print("HIGH SCORE TRACKER")
    print("Enter your scores. Type -1 when you're finished.")

    # Loop to collect scores from user until they enter -1
    while True:
        try:
            # Get input from user
            user_input = input("Enter a score (-1 to quit): ")
            score = float(user_input)

            # Check for sentinel value to exit
            if score == -1:
                break

            # Validate score (assuming scores should be non-negative)
            if score < 0:
                print("Please enter a non-negative score or -1 to quit.")
                continue

            # Add valid score to list
            scores.append(score)
            print(f"Score {score} added!")

        except ValueError:
            print("Invalid input. Please enter a number.")

    # Sort scores in descending order
    scores.sort(reverse=True)

    # Get top 10 scores (or all if fewer than 10)
    top_scores = scores[:10]

    # Print top scores to screen
    print("\nTop Scores:")
    if not top_scores:
        print("No scores entered.")
    else:
        for i, score in enumerate(top_scores, 1):
            print(f"{i}. {score}")

        # Write top scores to file
        with open("highscores.txt", "w") as file:
            for i, score in enumerate(top_scores, 1):
                file.write(f"{i}. {score}\n")

        print("\nScores have been saved to highscores.txt")

if __name__ == "__main__":
    main()3437W1
```

## **Part B** - Data Permanence

Here we will add the second part of retaining a highscores list - being able to read them back into the program.<br/><br/>

### Your Task

Copy your code from Part A and paste it into the code block below.<br/><br/>
Next, you need to extend the program to:
* open the `highscores.txt` file
* read and print the current highscores list to the screen
* populate the highscores list<br/><br/>

Once this is done, you need to ensure that the functionality in Part A still works - that is, that you can take in scores via the loop and add them to the list, and then re-save them to the file.

### Your Answer

```python
import datetime
import json
import csv
import statistics

def load_highscores():
    """Load highscores from file and return as list of dictionaries"""
    highscores = []
    try:
        with open("highscores.txt", "r") as file:
            for line in file:
                # Parse line with format: "1. 100.0 - Player (YYYY-MM-DD HH:MM:SS)"
                parts = line.strip().split(' - ')
                if len(parts) >= 2:
                    score_part = parts[0]
                    # Extract score number
                    score_num = score_part.split('. ', 1)[1] if '. ' in score_part else score_part

                    # Extract player name and datetime
                    player_time_part = parts[1]
                    player_time_parts = player_time_part.split(' (', 1)
                    player = player_time_parts[0]

                    # Extract datetime if it exists
                    timestamp = None
                    if len(player_time_parts) > 1 and ')' in player_time_parts[1]:
                        timestamp = player_time_parts[1].rstrip(')')

                    try:
                        score = float(score_num)
                        highscores.append({
                            'score': score,
                            'player': player,
                            'timestamp': timestamp
                        })
                    except ValueError:
                        # Skip invalid lines
                        pass
            print(f"Loaded {len(highscores)} highscores from file.")
    except FileNotFoundError:
        print("No existing high scores found. Starting fresh!")
    except Exception as e:
        print(f"Error reading highscores file: {e}")
    return highscores

def save_highscores(highscores):
    """Save highscores list to file"""
    # Sort before saving
    highscores.sort(key=lambda x: x['score'], reverse=True)
    top_scores = highscores[:10]  # Keep only top 10

    try:
        with open("highscores.txt", "w") as file:
            for i, entry in enumerate(top_scores, 1):
                # Format: "1. 100.0 - Player (YYYY-MM-DD HH:MM:SS)"
                timestamp_str = f" ({entry['timestamp']})" if entry['timestamp'] else ""
                file.write(f"{i}. {entry['score']} - {entry['player']}{timestamp_str}\n")
        print("\nScores have been saved to highscores.txt")
        return True
    except Exception as e:
        print(f"Error saving highscores: {e}")
        return False

def display_highscores(highscores):
    """Display highscores on screen"""
    if not highscores:
        print("No highscores available.")
        return

    print("\n===== HIGHSCORES =====")
    # Sort scores in descending order
    sorted_scores = sorted(highscores, key=lambda x: x['score'], reverse=True)
    for i, entry in enumerate(sorted_scores[:10], 1):  # Show top 10
        timestamp_str = f" ({entry['timestamp']})" if entry['timestamp'] else ""
        print(f"{i}. {entry['score']} - {entry['player']}{timestamp_str}")
    print("=====================\n")

def add_score(highscores):
    """Add a new score to the highscores list"""
    try:
        score = float(input("Enter the score: "))
        if score < 0:
            print("Score must be non-negative.")
            return highscores

        player = input("Enter player name: ")
        if not player.strip():
            player = "Anonymous"

        # Get current timestamp
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        # Add to highscores
        highscores.append({
            'score': score,
            'player': player,
            'timestamp': timestamp
        })

        print(f"Score {score} for {player} added!")
        return highscores
    except ValueError:
        print("Invalid score. Please enter a number.")
        return highscores

def add_multiple_scores(highscores):
    """Add multiple scores until a negative number is entered"""
    print("\n===== ADD MULTIPLE SCORES =====")
    print("Enter scores one by one. Enter a negative number to finish and return to the main menu.")

    scores_added = 0
    new_entries = []  # Track new entries for display

    while True:
        try:
            score_input = input(f"Enter score #{scores_added + 1} (negative number to finish): ")
            if not score_input.strip():
                print("No input provided. Try again or enter a negative number to finish.")
                continue

            score = float(score_input)
            if score < 0:
                break

            player = input("Enter player name: ")
            if not player.strip():
                player = "Anonymous"

            # Get current timestamp
            timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

            # Create score entry
            entry = {
                'score': score,
                'player': player,
                'timestamp': timestamp
            }

            # Add to highscores and track in new entries
            highscores.append(entry)
            new_entries.append(entry)

            print(f"Score {score} for {player} added!")
            scores_added += 1

        except ValueError:
            print("Invalid score. Please enter a number.")

    if scores_added > 0:
        print(f"\n{scores_added} new scores added successfully:")

        # Display the newly added scores
        print("\n=== NEWLY ADDED SCORES ===")
        for i, entry in enumerate(new_entries, 1):
            print(f"{i}. {entry['score']} - {entry['player']} ({entry['timestamp']})")
        print("==========================")

        # Prompt to save immediately or continue
        save_choice = input("Save these scores now? (y/n): ").lower()
        if save_choice == 'y':
            save_highscores(highscores)
    else:
        print("\nNo new scores were added.")

    return highscores

def delete_score(highscores):
    """Delete a score from the highscores list"""
    if not highscores:
        print("No highscores to delete.")
        return highscores

    display_highscores(highscores)
    try:
        index = int(input("Enter the number of the score to delete (0 to cancel): "))
        if index == 0:
            return highscores

        if 1 <= index <= len(highscores):
            # Use index-1 because the displayed list starts from 1
            sorted_scores = sorted(highscores, key=lambda x: x['score'], reverse=True)
            score_to_delete = sorted_scores[index-1]

            # Find and remove the score from the original list
            for i, score in enumerate(highscores):
                if (score['score'] == score_to_delete['score'] and
                    score['player'] == score_to_delete['player'] and
                    score['timestamp'] == score_to_delete['timestamp']):
                    del highscores[i]
                    print(f"Score #{index} deleted.")
                    break
        else:
            print("Invalid index.")
    except ValueError:
        print("Please enter a valid number.")

    return highscores

def modify_score(highscores):
    """Modify an existing score"""
    if not highscores:
        print("No highscores to modify.")
        return highscores

    display_highscores(highscores)
    try:
        index = int(input("Enter the number of the score to modify (0 to cancel): "))
        if index == 0:
            return highscores

        if 1 <= index <= len(highscores):
            # Use index-1 because the displayed list starts from 1
            sorted_scores = sorted(highscores, key=lambda x: x['score'], reverse=True)
            score_to_modify = sorted_scores[index-1]

            # Find the score in the original list
            for i, score in enumerate(highscores):
                if (score['score'] == score_to_modify['score'] and
                    score['player'] == score_to_modify['player'] and
                    score['timestamp'] == score_to_modify['timestamp']):

                    # Get new values
                    try:
                        new_score = float(input(f"Enter new score (current: {score['score']}): "))
                        if new_score < 0:
                            print("Score must be non-negative.")
                            return highscores
                    except ValueError:
                        print("Invalid score. Modification cancelled.")
                        return highscores

                    new_player = input(f"Enter new player name (current: {score['player']}, press Enter to keep): ")
                    if not new_player.strip():
                        new_player = score['player']

                    # Update values
                    highscores[i]['score'] = new_score
                    highscores[i]['player'] = new_player
                    # Update timestamp to current time
                    highscores[i]['timestamp'] = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

                    print(f"Score #{index} modified.")
                    break
        else:
            print("Invalid index.")
    except ValueError:
        print("Please enter a valid number.")

    return highscores

def clear_highscores():
    """Clear the entire highscores list"""
    confirm = input("Are you sure you want to clear all highscores? (y/n): ").lower()
    if confirm == 'y':
        return []
    else:
        print("Operation cancelled.")
        return None  # Indicates no change

def visualize_scores_text(highscores):
    """Create text-based visualizations of highscores"""
    if not highscores:
        print("No data to visualize.")
        return

    print("\n===== DATA VISUALIZATION =====")
    print("1. Score Distribution Histogram")
    print("2. Top Players Bar Chart")
    print("3. Score Timeline")
    print("4. Back to Main Menu")
    print("============================")

    choice = input("Select visualization type (1-4): ")

    if choice == '1':
        # Text-based score distribution histogram
        scores = [entry['score'] for entry in highscores]
        if not scores:
            print("No scores available to visualize.")
            return

        # Find min and max for scale
        min_score = min(scores)
        max_score = max(scores)

        # Create bins
        num_bins = 10
        if max_score - min_score <= 0:  # All scores are the same
            bin_width = 1
        else:
            bin_width = (max_score - min_score) / num_bins

        bins = [0] * num_bins
        for score in scores:
            if score == max_score:  # Handle edge case for max value
                bin_index = num_bins - 1
            else:
                bin_index = int((score - min_score) / bin_width)
            bins[bin_index] += 1

        # Find maximum bin height for scaling
        max_bin_height = max(bins) if bins else 0

        # Display histogram
        print("\n===== SCORE DISTRIBUTION HISTOGRAM =====")
        histogram_width = 40  # Character width of the histogram

        for i, count in enumerate(bins):
            if max_bin_height > 0:
                bar_length = int((count / max_bin_height) * histogram_width)
            else:
                bar_length = 0

            lower_bound = min_score + (i * bin_width)
            upper_bound = min_score + ((i + 1) * bin_width)
            print(f"{lower_bound:6.1f} - {upper_bound:6.1f} | {'#' * bar_length} ({count})")

        print("=========================================")

    elif choice == '2':
        # Text-based top players bar chart
        # Group scores by player and get their best score
        player_scores = {}
        for entry in highscores:
            player = entry['player']
            score = entry['score']
            if player not in player_scores or score > player_scores[player]:
                player_scores[player] = score

        # Sort by score and get top 5
        top_players = sorted(player_scores.items(), key=lambda x: x[1], reverse=True)[:5]

        if not top_players:
            print("Not enough player data for visualization.")
            return

        # Find max score for scaling
        max_score = max([score for _, score in top_players])

        # Display bar chart
        print("\n===== TOP 5 PLAYERS BAR CHART =====")
        chart_width = 40  # Character width of the chart

        for player, score in top_players:
            bar_length = int((score / max_score) * chart_width) if max_score > 0 else 0
            print(f"{player:15} | {'#' * bar_length} ({score:.1f})")

        print("==================================")

    elif choice == '3':
        # Text-based score timeline
        # Filter entries with valid timestamps
        valid_entries = []
        for entry in highscores:
            if entry['timestamp']:
                try:
                    # Convert timestamp string to datetime object
                    dt = datetime.datetime.strptime(entry['timestamp'], "%Y-%m-%d %H:%M:%S")
                    valid_entries.append({
                        'score': entry['score'],
                        'player': entry['player'],
                        'datetime': dt
                    })
                except ValueError:
                    # Skip entries with invalid timestamp format
                    pass

        if not valid_entries:
            print("No valid timestamp data for timeline visualization.")
            return

        # Sort by datetime
        valid_entries.sort(key=lambda x: x['datetime'])

        # Group by date to simplify text display
        date_scores = {}
        for entry in valid_entries:
            date_str = entry['datetime'].strftime("%Y-%m-%d")
            if date_str not in date_scores:
                date_scores[date_str] = []
            date_scores[date_str].append((entry['score'], entry['player']))

        # Display timeline
        print("\n===== SCORE TIMELINE =====")
        for date_str in sorted(date_scores.keys()):
            print(f"\n{date_str}:")
            for score, player in date_scores[date_str]:
                print(f"  {player}: {score:.1f}")

        print("==========================")

    elif choice == '4':
        return
    else:
        print("Invalid choice.")

def export_import_menu(highscores):
    """Menu for export and import operations"""
    print("\n===== EXPORT/IMPORT MENU =====")
    print("1. Export to CSV")
    print("2. Export to JSON")
    print("3. Import from CSV")
    print("4. Import from JSON")
    print("5. Back to Main Menu")
    print("============================")

    choice = input("Select operation (1-5): ")

    if choice == '1':
        return export_to_csv(highscores)
    elif choice == '2':
        return export_to_json(highscores)
    elif choice == '3':
        result = import_from_csv()
        return result if result is not None else highscores
    elif choice == '4':
        result = import_from_json()
        return result if result is not None else highscores
    elif choice == '5':
        return highscores
    else:
        print("Invalid choice.")
        return highscores

def export_to_csv(highscores):
    """Export highscores to CSV file"""
    if not highscores:
        print("No data to export.")
        return highscores

    filename = input("Enter filename for CSV export (default: highscores.csv): ")
    if not filename.strip():
        filename = "highscores.csv"
    if not filename.endswith('.csv'):
        filename += '.csv'

    try:
        with open(filename, 'w', newline='') as csvfile:
            fieldnames = ['score', 'player', 'timestamp']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

            writer.writeheader()
            for entry in highscores:
                writer.writerow(entry)

        print(f"Successfully exported {len(highscores)} records to {filename}")
    except Exception as e:
        print(f"Error exporting to CSV: {e}")

    return highscores

def export_to_json(highscores):
    """Export highscores to JSON file"""
    if not highscores:
        print("No data to export.")
        return highscores

    filename = input("Enter filename for JSON export (default: highscores.json): ")
    if not filename.strip():
        filename = "highscores.json"
    if not filename.endswith('.json'):
        filename += '.json'

    try:
        with open(filename, 'w') as jsonfile:
            json.dump(highscores, jsonfile, indent=4)

        print(f"Successfully exported {len(highscores)} records to {filename}")
    except Exception as e:
        print(f"Error exporting to JSON: {e}")

    return highscores

def import_from_csv():
    """Import highscores from CSV file"""
    filename = input("Enter CSV filename to import (default: highscores.csv): ")
    if not filename.strip():
        filename = "highscores.csv"

    imported_scores = []
    try:
        with open(filename, 'r', newline='') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                # Validate data
                try:
                    score = float(row['score'])
                    player = row['player'] if 'player' in row and row['player'] else "Anonymous"
                    timestamp = row['timestamp'] if 'timestamp' in row else None

                    imported_scores.append({
                        'score': score,
                        'player': player,
                        'timestamp': timestamp
                    })
                except (ValueError, KeyError):
                    # Skip invalid rows
                    continue

        if imported_scores:
            print(f"Successfully imported {len(imported_scores)} records from {filename}")

            action = input("Do you want to (A)ppend to current scores or (R)eplace them? (A/R): ").upper()
            if action == 'R':
                return imported_scores
            else:  # Default to append
                confirm = input(f"Append {len(imported_scores)} records to existing list? (y/n): ").lower()
                if confirm == 'y':
                    current_scores = load_highscores()
                    return current_scores + imported_scores
                else:
                    print("Import cancelled.")
                    return None
        else:
            print("No valid records found in the file.")
            return None
    except FileNotFoundError:
        print(f"File {filename} not found.")
        return None
    except Exception as e:
        print(f"Error importing from CSV: {e}")
        return None

def import_from_json():
    """Import highscores from JSON file"""
    filename = input("Enter JSON filename to import (default: highscores.json): ")
    if not filename.strip():
        filename = "highscores.json"

    try:
        with open(filename, 'r') as jsonfile:
            imported_data = json.load(jsonfile)

        if not isinstance(imported_data, list):
            print("Invalid JSON format. Expected a list of score records.")
            return None

        # Validate and clean data
        imported_scores = []
        for item in imported_data:
            if isinstance(item, dict) and 'score' in item:
                try:
                    score = float(item['score'])
                    player = item.get('player', "Anonymous")
                    timestamp = item.get('timestamp')

                    imported_scores.append({
                        'score': score,
                        'player': player,
                        'timestamp': timestamp
                    })
                except (ValueError, TypeError):
                    # Skip invalid items
                    continue

        if imported_scores:
            print(f"Successfully imported {len(imported_scores)} records from {filename}")

            action = input("Do you want to (A)ppend to current scores or (R)eplace them? (A/R): ").upper()
            if action == 'R':
                return imported_scores
            else:  # Default to append
                confirm = input(f"Append {len(imported_scores)} records to existing list? (y/n): ").lower()
                if confirm == 'y':
                    current_scores = load_highscores()
                    return current_scores + imported_scores
                else:
                    print("Import cancelled.")
                    return None
        else:
            print("No valid records found in the file.")
            return None
    except FileNotFoundError:
        print(f"File {filename} not found.")
        return None
    except json.JSONDecodeError:
        print(f"Invalid JSON format in {filename}.")
        return None
    except Exception as e:
        print(f"Error importing from JSON: {e}")
        return None

def show_statistics(highscores):
    """Display advanced statistics about the highscores"""
    if not highscores:
        print("No data available for statistics.")
        return

    scores = [entry['score'] for entry in highscores]

    print("\n===== SCORE STATISTICS =====")

    # Basic statistics
    print(f"Total number of scores: {len(scores)}")
    print(f"Highest score: {max(scores)}")
    print(f"Lowest score: {min(scores)}")
    print(f"Average score: {sum(scores)/len(scores):.2f}")

    # Add median and mode if we have enough data
    if len(scores) >= 3:
        print(f"Median score: {statistics.median(scores):.2f}")
        if len(scores) >= 5:  # Mode makes more sense with more data points
            try:
                print(f"Mode score: {statistics.mode(scores):.2f}")
            except statistics.StatisticsError:
                # No unique mode
                print("Mode: No single most common score")

    # Standard deviation
    if len(scores) >= 2:  # Need at least 2 scores for std dev
        print(f"Standard deviation: {statistics.stdev(scores):.2f}")

    print("\n===== PLAYER STATISTICS =====")

    # Player statistics - using regular dict instead of defaultdict
    player_scores = {}
    for entry in highscores:
        player = entry['player']
        if player not in player_scores:
            player_scores[player] = []
        player_scores[player].append(entry['score'])

    # Number of unique players
    print(f"Number of players: {len(player_scores)}")

    # Top players by average score (with at least 2 scores)
    print("\nTop players by average score (min 2 scores):")
    avg_scores = []
    for player, scores_list in player_scores.items():
        if len(scores_list) >= 2:
            avg = sum(scores_list) / len(scores_list)
            avg_scores.append((player, avg, len(scores_list)))

    if avg_scores:
        avg_scores.sort(key=lambda x: x[1], reverse=True)
        for i, (player, avg, count) in enumerate(avg_scores[:5], 1):
            print(f"{i}. {player}: {avg:.2f} (from {count} scores)")
    else:
        print("No players with multiple scores.")

    # Most consistent player (lowest standard deviation)
    print("\nMost consistent players (min 3 scores):")
    consistency = []
    for player, scores_list in player_scores.items():
        if len(scores_list) >= 3:
            stdev = statistics.stdev(scores_list)
            consistency.append((player, stdev, len(scores_list)))

    if consistency:
        consistency.sort(key=lambda x: x[1])  # Sort by lowest stdev
        for i, (player, stdev, count) in enumerate(consistency[:3], 1):
            print(f"{i}. {player}: σ={stdev:.2f} (from {count} scores)")
    else:
        print("No players with 3+ scores for consistency analysis.")

    # Time-based statistics
    print("\n===== TIME-BASED STATISTICS =====")

    # Parse timestamps
    valid_timestamps = []
    for entry in highscores:
        if entry['timestamp']:
            try:
                dt = datetime.datetime.strptime(entry['timestamp'], "%Y-%m-%d %H:%M:%S")
                valid_timestamps.append((dt, entry['score'], entry['player']))
            except ValueError:
                pass

    if valid_timestamps:
        # Sort by timestamp
        valid_timestamps.sort(key=lambda x: x[0])

        # First and last recorded score
        first_time, first_score, first_player = valid_timestamps[0]
        last_time, last_score, last_player = valid_timestamps[-1]

        print(f"First recorded score: {first_score} by {first_player} on {first_time.strftime('%Y-%m-%d')}")
        print(f"Most recent score: {last_score} by {last_player} on {last_time.strftime('%Y-%m-%d')}")

        # Group by month if enough data
        if len(valid_timestamps) >= 5 and (last_time - first_time).days >= 30:
            print("\nAverage score by month:")
            month_scores = {}

            for dt, score, _ in valid_timestamps:
                month_key = dt.strftime("%Y-%m")
                if month_key not in month_scores:
                    month_scores[month_key] = []
                month_scores[month_key].append(score)

            for month, month_score_list in sorted(month_scores.items()):
                avg = sum(month_score_list) / len(month_score_list)
                print(f"{month}: {avg:.2f} (from {len(month_score_list)} scores)")
    else:
        print("No valid timestamp data for time-based statistics.")

    print("=============================")
    input("Press Enter to continue...")

def display_menu():
    """Display the main menu options"""
    print("\n===== HIGH SCORE TRACKER =====")
    print("1. Add single score")
    print("2. Add multiple scores")
    print("3. View scores")
    print("4. Delete score")
    print("5. Modify score")
    print("6. Clear highscore list")
    print("7. Data Visualization")
    print("8. Export/Import")
    print("9. Advanced Statistics")
    print("10. Exit & Save")
    print("============================")

def main():
    # Load existing highscores
    highscores = load_highscores()

    while True:
        display_menu()
        choice = input("Enter your choice (1-10): ")

        if choice == '1':
            highscores = add_score(highscores)
        elif choice == '2':
            highscores = add_multiple_scores(highscores)
        elif choice == '3':
            display_highscores(highscores)
        elif choice == '4':
            highscores = delete_score(highscores)
        elif choice == '5':
            highscores = modify_score(highscores)
        elif choice == '6':
            result = clear_highscores()
            if result is not None:  # Check if operation wasn't cancelled
                highscores = result
        elif choice == '7':
            visualize_scores_text(highscores)
        elif choice == '8':
            highscores = export_import_menu(highscores)
        elif choice == '9':
            show_statistics(highscores)
        elif choice == '10':
            save_highscores(highscores)
            print("Thank you for using High Score Tracker!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 10.")

if __name__ == "__main__":
    main()
```

```python
# Student Name: Ryan Cushen
# Student Number: 1035373
# The visualization system now offers three different chart types:
# Histogram showing the distribution of all scores
# Bar chart of the top players
# Timeline showing score progression over time
# The program detect the OS and then will automatically open visualisation in default image viewer
# Import scores from file and export scores to file
  # Multiple Formats: Support for both CSV and JSON formats
  # Flexible Importing: Option to append to or replace existing scores
  # Custom Filenames: Users can specify custom filenames for exports
  # Data Validation: Thorough validation to ensure only valid data is imported
# Score trends over time including monthly averages

# This enhanced version requires these Python libraries:

  # matplotlib for creating visualizations
  # statistics for advanced statistical calculations
  # Standard library modules: json, csv, datetime, collections

import os
import datetime
import json
import csv
import statistics
import webbrowser
import subprocess
import platform
from collections import defaultdict
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import numpy as np

def load_highscores():
    """Load highscores from file and return as list of dictionaries"""
    highscores = []
    if os.path.exists("highscores.txt"):
        try:
            with open("highscores.txt", "r") as file:
                for line in file:
                    # Parse line with format: "1. 100.0 - Player (YYYY-MM-DD HH:MM:SS)"
                    parts = line.strip().split(' - ')
                    if len(parts) >= 2:
                        score_part = parts[0]
                        # Extract score number
                        score_num = score_part.split('. ', 1)[1] if '. ' in score_part else score_part

                        # Extract player name and datetime
                        player_time_part = parts[1]
                        player_time_parts = player_time_part.split(' (', 1)
                        player = player_time_parts[0]

                        # Extract datetime if it exists
                        timestamp = None
                        if len(player_time_parts) > 1 and ')' in player_time_parts[1]:
                            timestamp = player_time_parts[1].rstrip(')')

                        try:
                            score = float(score_num)
                            highscores.append({
                                'score': score,
                                'player': player,
                                'timestamp': timestamp
                            })
                        except ValueError:
                            # Skip invalid lines
                            pass
            print(f"Loaded {len(highscores)} highscores from file.")
        except Exception as e:
            print(f"Error reading highscores file: {e}")
    return highscores

def save_highscores(highscores):
    """Save highscores list to file"""
    # Sort before saving
    highscores.sort(key=lambda x: x['score'], reverse=True)
    top_scores = highscores[:10]  # Keep only top 10

    try:
        with open("highscores.txt", "w") as file:
            for i, entry in enumerate(top_scores, 1):
                # Format: "1. 100.0 - Player (YYYY-MM-DD HH:MM:SS)"
                timestamp_str = f" ({entry['timestamp']})" if entry['timestamp'] else ""
                file.write(f"{i}. {entry['score']} - {entry['player']}{timestamp_str}\n")
        print("\nScores have been saved to highscores.txt")
        return True
    except Exception as e:
        print(f"Error saving highscores: {e}")
        return False

def display_highscores(highscores):
    """Display highscores on screen"""
    if not highscores:
        print("No highscores available.")
        return

    print("\n===== HIGHSCORES =====")
    # Sort scores in descending order
    sorted_scores = sorted(highscores, key=lambda x: x['score'], reverse=True)
    for i, entry in enumerate(sorted_scores[:10], 1):  # Show top 10
        timestamp_str = f" ({entry['timestamp']})" if entry['timestamp'] else ""
        print(f"{i}. {entry['score']} - {entry['player']}{timestamp_str}")
    print("=====================\n")

def add_score(highscores):
    """Add a new score to the highscores list"""
    try:
        score = float(input("Enter the score: "))
        if score < 0:
            print("Score must be non-negative.")
            return highscores

        player = input("Enter player name: ")
        if not player.strip():
            player = "Anonymous"

        # Get current timestamp
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        # Add to highscores
        highscores.append({
            'score': score,
            'player': player,
            'timestamp': timestamp
        })

        print(f"Score {score} for {player} added!")
        return highscores
    except ValueError:
        print("Invalid score. Please enter a number.")
        return highscores

def delete_score(highscores):
    """Delete a score from the highscores list"""
    if not highscores:
        print("No highscores to delete.")
        return highscores

    display_highscores(highscores)
    try:
        index = int(input("Enter the number of the score to delete (0 to cancel): "))
        if index == 0:
            return highscores

        if 1 <= index <= len(highscores):
            # Use index-1 because the displayed list starts from 1
            sorted_scores = sorted(highscores, key=lambda x: x['score'], reverse=True)
            score_to_delete = sorted_scores[index-1]

            # Find and remove the score from the original list
            for i, score in enumerate(highscores):
                if (score['score'] == score_to_delete['score'] and
                    score['player'] == score_to_delete['player'] and
                    score['timestamp'] == score_to_delete['timestamp']):
                    del highscores[i]
                    print(f"Score #{index} deleted.")
                    break
        else:
            print("Invalid index.")
    except ValueError:
        print("Please enter a valid number.")

    return highscores

def modify_score(highscores):
    """Modify an existing score"""
    if not highscores:
        print("No highscores to modify.")
        return highscores

    display_highscores(highscores)
    try:
        index = int(input("Enter the number of the score to modify (0 to cancel): "))
        if index == 0:
            return highscores

        if 1 <= index <= len(highscores):
            # Use index-1 because the displayed list starts from 1
            sorted_scores = sorted(highscores, key=lambda x: x['score'], reverse=True)
            score_to_modify = sorted_scores[index-1]

            # Find the score in the original list
            for i, score in enumerate(highscores):
                if (score['score'] == score_to_modify['score'] and
                    score['player'] == score_to_modify['player'] and
                    score['timestamp'] == score_to_modify['timestamp']):

                    # Get new values
                    try:
                        new_score = float(input(f"Enter new score (current: {score['score']}): "))
                        if new_score < 0:
                            print("Score must be non-negative.")
                            return highscores
                    except ValueError:
                        print("Invalid score. Modification cancelled.")
                        return highscores

                    new_player = input(f"Enter new player name (current: {score['player']}, press Enter to keep): ")
                    if not new_player.strip():
                        new_player = score['player']

                    # Update values
                    highscores[i]['score'] = new_score
                    highscores[i]['player'] = new_player
                    # Update timestamp to current time
                    highscores[i]['timestamp'] = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

                    print(f"Score #{index} modified.")
                    break
        else:
            print("Invalid index.")
    except ValueError:
        print("Please enter a valid number.")

    return highscores

def clear_highscores():
    """Clear the entire highscores list"""
    confirm = input("Are you sure you want to clear all highscores? (y/n): ").lower()
    if confirm == 'y':
        return []
    else:
        print("Operation cancelled.")
        return None  # Indicates no change

# NEW FEATURE: Data Visualization
def open_image_file(filename):
    """Open an image file using the system's default viewer"""
    try:
        # Determine the operating system and use appropriate method
        system = platform.system()

        if system == 'Darwin':  # macOS
            subprocess.call(['open', filename])
        elif system == 'Windows':
            # On Windows, webbrowser.open works well for images
            webbrowser.open(filename)
        else:  # Linux and others
            # Try xdg-open for Linux first
            try:
                subprocess.call(['xdg-open', filename])
            except:
                # Fall back to webbrowser if xdg-open fails
                webbrowser.open(filename)

        print(f"Opening {filename} in default image viewer...")
        return True
    except Exception as e:
        print(f"Error opening image file: {e}")
        print(f"The image was saved to {filename} but couldn't be opened automatically.")
        return False

def visualize_scores(highscores):
    """Create visual charts of highscores"""
    if not highscores:
        print("No data to visualize.")
        return

    print("\n===== DATA VISUALIZATION =====")
    print("1. Score Distribution")
    print("2. Top Players Bar Chart")
    print("3. Score Timeline")
    print("4. Back to Main Menu")
    print("============================")

    choice = input("Select visualization type (1-4): ")

    if choice == '1':
        # Score Distribution Histogram
        plt.figure(figsize=(10, 6))
        scores = [entry['score'] for entry in highscores]
        plt.hist(scores, bins=10, alpha=0.7, color='skyblue', edgecolor='black')
        plt.title('Score Distribution Histogram')
        plt.xlabel('Score Value')
        plt.ylabel('Frequency')
        plt.grid(axis='y', alpha=0.75)
        filename = 'score_distribution.png'
        plt.savefig(filename)
        plt.close()
        print(f"Score distribution histogram saved as '{filename}'")
        open_image_file(filename)

    elif choice == '2':
        # Top Players Bar Chart
        # Group scores by player and get their best score
        player_scores = {}
        for entry in highscores:
            player = entry['player']
            score = entry['score']
            if player not in player_scores or score > player_scores[player]:
                player_scores[player] = score

        # Sort by score and get top 5
        top_players = sorted(player_scores.items(), key=lambda x: x[1], reverse=True)[:5]

        if not top_players:
            print("Not enough player data for visualization.")
            return

        players = [p[0] for p in top_players]
        scores = [p[1] for p in top_players]

        plt.figure(figsize=(10, 6))
        plt.bar(players, scores, color='skyblue')
        plt.title('Top 5 Players by Best Score')
        plt.xlabel('Player')
        plt.ylabel('Best Score')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        filename = 'top_players.png'
        plt.savefig(filename)
        plt.close()
        print(f"Top players chart saved as '{filename}'")
        open_image_file(filename)

    elif choice == '3':
        # Score Timeline Chart
        # Filter entries with valid timestamps
        valid_entries = []
        for entry in highscores:
            if entry['timestamp']:
                try:
                    # Convert timestamp string to datetime object
                    dt = datetime.datetime.strptime(entry['timestamp'], "%Y-%m-%d %H:%M:%S")
                    valid_entries.append({
                        'score': entry['score'],
                        'player': entry['player'],
                        'datetime': dt
                    })
                except ValueError:
                    # Skip entries with invalid timestamp format
                    pass

        if not valid_entries:
            print("No valid timestamp data for timeline visualization.")
            return

        # Sort by datetime
        valid_entries.sort(key=lambda x: x['datetime'])

        # Prepare data for plotting
        dates = [entry['datetime'] for entry in valid_entries]
        scores = [entry['score'] for entry in valid_entries]
        players = [entry['player'] for entry in valid_entries]

        plt.figure(figsize=(12, 6))

        # Plot with different markers/colors per player
        player_set = set(players)
        color_map = plt.cm.get_cmap('tab10', len(player_set))
        colors = {player: color_map(i) for i, player in enumerate(player_set)}

        for player in player_set:
            player_dates = [dates[i] for i in range(len(dates)) if players[i] == player]
            player_scores = [scores[i] for i in range(len(scores)) if players[i] == player]
            plt.plot(player_dates, player_scores, 'o-', label=player, color=colors[player])

        plt.title('Score Timeline')
        plt.xlabel('Date')
        plt.ylabel('Score')
        plt.legend()
        plt.grid(True, alpha=0.3)

        # Format x-axis dates
        plt.gcf().autofmt_xdate()
        date_format = DateFormatter("%Y-%m-%d")
        plt.gca().xaxis.set_major_formatter(date_format)

        plt.tight_layout()
        filename = 'score_timeline.png'
        plt.savefig(filename)
        plt.close()
        print(f"Score timeline chart saved as '{filename}'")
        open_image_file(filename)

    elif choice == '4':
        return
    else:
        print("Invalid choice.")

# NEW FEATURE: Export/Import
def export_import_menu(highscores):
    """Menu for export and import operations"""
    print("\n===== EXPORT/IMPORT MENU =====")
    print("1. Export to CSV")
    print("2. Export to JSON")
    print("3. Import from CSV")
    print("4. Import from JSON")
    print("5. Back to Main Menu")
    print("============================")

    choice = input("Select operation (1-5): ")

    if choice == '1':
        return export_to_csv(highscores)
    elif choice == '2':
        return export_to_json(highscores)
    elif choice == '3':
        result = import_from_csv()
        return result if result is not None else highscores
    elif choice == '4':
        result = import_from_json()
        return result if result is not None else highscores
    elif choice == '5':
        return highscores
    else:
        print("Invalid choice.")
        return highscores

def export_to_csv(highscores):
    """Export highscores to CSV file"""
    if not highscores:
        print("No data to export.")
        return highscores

    filename = input("Enter filename for CSV export (default: highscores.csv): ")
    if not filename.strip():
        filename = "highscores.csv"
    if not filename.endswith('.csv'):
        filename += '.csv'

    try:
        with open(filename, 'w', newline='') as csvfile:
            fieldnames = ['score', 'player', 'timestamp']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

            writer.writeheader()
            for entry in highscores:
                writer.writerow(entry)

        print(f"Successfully exported {len(highscores)} records to {filename}")
    except Exception as e:
        print(f"Error exporting to CSV: {e}")

    return highscores

def export_to_json(highscores):
    """Export highscores to JSON file"""
    if not highscores:
        print("No data to export.")
        return highscores

    filename = input("Enter filename for JSON export (default: highscores.json): ")
    if not filename.strip():
        filename = "highscores.json"
    if not filename.endswith('.json'):
        filename += '.json'

    try:
        with open(filename, 'w') as jsonfile:
            json.dump(highscores, jsonfile, indent=4)

        print(f"Successfully exported {len(highscores)} records to {filename}")
    except Exception as e:
        print(f"Error exporting to JSON: {e}")

    return highscores

def import_from_csv():
    """Import highscores from CSV file"""
    filename = input("Enter CSV filename to import (default: highscores.csv): ")
    if not filename.strip():
        filename = "highscores.csv"

    if not os.path.exists(filename):
        print(f"File {filename} not found.")
        return None

    imported_scores = []
    try:
        with open(filename, 'r', newline='') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                # Validate data
                try:
                    score = float(row['score'])
                    player = row['player'] if 'player' in row and row['player'] else "Anonymous"
                    timestamp = row['timestamp'] if 'timestamp' in row else None

                    imported_scores.append({
                        'score': score,
                        'player': player,
                        'timestamp': timestamp
                    })
                except (ValueError, KeyError):
                    # Skip invalid rows
                    continue

        if imported_scores:
            print(f"Successfully imported {len(imported_scores)} records from {filename}")

            action = input("Do you want to (A)ppend to current scores or (R)eplace them? (A/R): ").upper()
            if action == 'R':
                return imported_scores
            else:  # Default to append
                confirm = input(f"Append {len(imported_scores)} records to existing list? (y/n): ").lower()
                if confirm == 'y':
                    current_scores = load_highscores()
                    return current_scores + imported_scores
                else:
                    print("Import cancelled.")
                    return None
        else:
            print("No valid records found in the file.")
            return None
    except Exception as e:
        print(f"Error importing from CSV: {e}")
        return None

def import_from_json():
    """Import highscores from JSON file"""
    filename = input("Enter JSON filename to import (default: highscores.json): ")
    if not filename.strip():
        filename = "highscores.json"

    if not os.path.exists(filename):
        print(f"File {filename} not found.")
        return None

    try:
        with open(filename, 'r') as jsonfile:
            imported_data = json.load(jsonfile)

        if not isinstance(imported_data, list):
            print("Invalid JSON format. Expected a list of score records.")
            return None

        # Validate and clean data
        imported_scores = []
        for item in imported_data:
            if isinstance(item, dict) and 'score' in item:
                try:
                    score = float(item['score'])
                    player = item.get('player', "Anonymous")
                    timestamp = item.get('timestamp')

                    imported_scores.append({
                        'score': score,
                        'player': player,
                        'timestamp': timestamp
                    })
                except (ValueError, TypeError):
                    # Skip invalid items
                    continue

        if imported_scores:
            print(f"Successfully imported {len(imported_scores)} records from {filename}")

            action = input("Do you want to (A)ppend to current scores or (R)eplace them? (A/R): ").upper()
            if action == 'R':
                return imported_scores
            else:  # Default to append
                confirm = input(f"Append {len(imported_scores)} records to existing list? (y/n): ").lower()
                if confirm == 'y':
                    current_scores = load_highscores()
                    return current_scores + imported_scores
                else:
                    print("Import cancelled.")
                    return None
        else:
            print("No valid records found in the file.")
            return None
    except Exception as e:
        print(f"Error importing from JSON: {e}")
        return None

# NEW FEATURE: Advanced Statistics
def show_statistics(highscores):
    """Display advanced statistics about the highscores"""
    if not highscores:
        print("No data available for statistics.")
        return

    scores = [entry['score'] for entry in highscores]

    print("\n===== SCORE STATISTICS =====")

    # Basic statistics
    print(f"Total number of scores: {len(scores)}")
    print(f"Highest score: {max(scores)}")
    print(f"Lowest score: {min(scores)}")
    print(f"Average score: {sum(scores)/len(scores):.2f}")

    # Add median and mode if we have enough data
    if len(scores) >= 3:
        print(f"Median score: {statistics.median(scores):.2f}")
        if len(scores) >= 5:  # Mode makes more sense with more data points
            try:
                print(f"Mode score: {statistics.mode(scores):.2f}")
            except statistics.StatisticsError:
                # No unique mode
                print("Mode: No single most common score")

    # Standard deviation
    if len(scores) >= 2:  # Need at least 2 scores for std dev
        print(f"Standard deviation: {statistics.stdev(scores):.2f}")

    print("\n===== PLAYER STATISTICS =====")

    # Player statistics
    player_scores = defaultdict(list)
    for entry in highscores:
        player_scores[entry['player']].append(entry['score'])

    # Number of unique players
    print(f"Number of players: {len(player_scores)}")

    # Top players by average score (with at least 2 scores)
    print("\nTop players by average score (min 2 scores):")
    avg_scores = []
    for player, scores_list in player_scores.items():
        if len(scores_list) >= 2:
            avg = sum(scores_list) / len(scores_list)
            avg_scores.append((player, avg, len(scores_list)))

    if avg_scores:
        avg_scores.sort(key=lambda x: x[1], reverse=True)
        for i, (player, avg, count) in enumerate(avg_scores[:5], 1):
            print(f"{i}. {player}: {avg:.2f} (from {count} scores)")
    else:
        print("No players with multiple scores.")

    # Most consistent player (lowest standard deviation)
    print("\nMost consistent players (min 3 scores):")
    consistency = []
    for player, scores_list in player_scores.items():
        if len(scores_list) >= 3:
            stdev = statistics.stdev(scores_list)
            consistency.append((player, stdev, len(scores_list)))

    if consistency:
        consistency.sort(key=lambda x: x[1])  # Sort by lowest stdev
        for i, (player, stdev, count) in enumerate(consistency[:3], 1):
            print(f"{i}. {player}: σ={stdev:.2f} (from {count} scores)")
    else:
        print("No players with 3+ scores for consistency analysis.")

    # Time-based statistics
    print("\n===== TIME-BASED STATISTICS =====")

    # Parse timestamps
    valid_timestamps = []
    for entry in highscores:
        if entry['timestamp']:
            try:
                dt = datetime.datetime.strptime(entry['timestamp'], "%Y-%m-%d %H:%M:%S")
                valid_timestamps.append((dt, entry['score'], entry['player']))
            except ValueError:
                pass

    if valid_timestamps:
        # Sort by timestamp
        valid_timestamps.sort(key=lambda x: x[0])

        # First and last recorded score
        first_time, first_score, first_player = valid_timestamps[0]
        last_time, last_score, last_player = valid_timestamps[-1]

        print(f"First recorded score: {first_score} by {first_player} on {first_time.strftime('%Y-%m-%d')}")
        print(f"Most recent score: {last_score} by {last_player} on {last_time.strftime('%Y-%m-%d')}")

        # Group by month if enough data
        if len(valid_timestamps) >= 5 and (last_time - first_time).days >= 30:
            print("\nAverage score by month:")
            month_scores = defaultdict(list)

            for dt, score, _ in valid_timestamps:
                month_key = dt.strftime("%Y-%m")
                month_scores[month_key].append(score)

            for month, month_score_list in sorted(month_scores.items()):
                avg = sum(month_score_list) / len(month_score_list)
                print(f"{month}: {avg:.2f} (from {len(month_score_list)} scores)")
    else:
        print("No valid timestamp data for time-based statistics.")

    print("=============================")
    input("Press Enter to continue...")

def display_menu():
    """Display the main menu options"""
    print("\n===== HIGH SCORE TRACKER =====")
    print("1. Add new score")
    print("2. View scores")
    print("3. Delete score")
    print("4. Modify score")
    print("5. Clear highscore list")
    print("6. Data Visualization")  # New option
    print("7. Export/Import")        # New option
    print("8. Advanced Statistics")  # New option
    print("9. Exit & Save")
    print("============================")

def main():
    # Load existing highscores
    highscores = load_highscores()

    while True:
        display_menu()
        choice = input("Enter your choice (1-9): ")

        if choice == '1':
            highscores = add_score(highscores)
        elif choice == '2':
            display_highscores(highscores)
        elif choice == '3':
            highscores = delete_score(highscores)
        elif choice == '4':
            highscores = modify_score(highscores)
        elif choice == '5':
            result = clear_highscores()
            if result is not None:  # Check if operation wasn't cancelled
                highscores = result
        elif choice == '6':
            visualize_scores(highscores)
        elif choice == '7':
            highscores = export_import_menu(highscores)
        elif choice == '8':
            show_statistics(highscores)
        elif choice == '9':
            save_highscores(highscores)
            print("Thank you for using High Score Tracker!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 9.")

if __name__ == "__main__":
    main()
```
